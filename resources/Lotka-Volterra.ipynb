```python
import matplotlib.pyplot as plt

r = 1.0
b = 1.0
d = 1.0
c = 1.0
K = 5.0
x = 1.0
y = 1.0
x_result = [x]
y_result = [y]
Steps = 300

for step in range(Steps):
    next_x = x + r*x*(1-x/K) - (1-1/(b*y+1))*x
    next_y = y - d*y + c*x*y
    x = next_x
    y = next_y
    if x < 0:
        x = 0
    if y < 0:
        y = 0
    x_result.append(x)
    y_result.append(y)
    
plt.plot(x_result,'r')
plt.plot(y_result,'b')
plt.show()
plt.plot(x_result,y_result)
plt.show()
```
